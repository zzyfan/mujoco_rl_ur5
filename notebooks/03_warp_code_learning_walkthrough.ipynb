## 2. 运行时检查层

`warp_ur5_runtime.py` 负责检查：
- `warp-lang`
- `mujoco-warp`
- `mujoco_playground`

并在训练前初始化 CUDA 设备。

```python
from warp_ur5_runtime import warp_available, playground_importable

print('warp_available =', warp_available())
print('playground_importable =', playground_importable())
```

## 3. 环境层

`warp_ur5_env.py` 使用：
- `jax`
- `jax.numpy`
- `mujoco`
- `mujoco.mjx`
- `mujoco_playground._src.mjx_env`

环境层的关键函数包括：
- `default_config()`
- `reset()`
- `step()`
- `_compute_action_torque()`
- `_compute_goal_reward()`


## 4. 训练层

`train_ur5_reach_warp.py` 不自己实现优化器，而是调用 Brax 的训练器：
- `brax.training.agents.sac.train`
- `brax.training.agents.ppo.train`

这个脚本的主要价值在于：
- 组织参数
- 构建环境
- 输出训练日志
- 保存配置和最终参数
