## 2. 构建环境并跑 5 步

Python 语法点：
- `from x import y` 直接导入类
- `for _ in range(n)` 表示执行 n 次且不关心计数变量

库用法：
- `gymnasium.Env` 规范了 `reset()` / `step()` / `render()` 接口
- `action_space.sample()` 从动作空间采样一个合法动作

```python
from ur5_reach_env import UR5ReachEnv

env = UR5ReachEnv(render_mode=None)
obs, info = env.reset()
print("obs shape:", obs.shape)

for _ in range(5):
    action = env.action_space.sample()
    obs, reward, done, truncated, info = env.step(action)
    print("reward:", reward, "done:", done, "truncated:", truncated)

env.close()
```

## 5. 常用参数速查

- `--algo`：选择算法（`td3` / `sac` / `ppo`）
- `--total-timesteps`：训练总步数
- `--n-envs`：并行环境数量
- `--model`：测试时选择 `best` 或 `final`
- `--render-mode` / `--render-every`：训练渲染设置
- `--device`：训练设备（`auto` / `cpu` / `cuda`）

## 6. 训练过程中的 Python 关键点

- `argparse`：解析命令行参数（把字符串转成正确类型）
- `if args.test:`：分支逻辑（训练/测试）
- `with` 不常出现，但在日志/文件操作时常用
