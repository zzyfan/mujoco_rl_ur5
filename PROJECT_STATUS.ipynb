# 项目状态笔记

这份笔记用于快速说明 `mujoco_ur5_rl` 当前的项目状态和使用入口。

- 主训练线：`classic/`
- GPU 训练线：`warp_gpu/`
- 支持的机器人：`ur5_cxy`、`zero_robotiq`


## 当前完成度

`classic/` 主线已经可以正常使用。

已验证内容：

- Python 源码可以正常通过语法检查。
- Gym 环境注册可以正常工作。
- `classic.test` 的随机策略 smoke test 在 `--physics-backend mujoco` 下可以运行。
- 当前默认物理后端是 `mujoco`。
- End-effector velocity defaults to finite-difference linear velocity.
- Legacy zeroarm velocity behavior can still be enabled with `--legacy-zero-ee-velocity`.


## 推荐命令

冒烟测试：

```bash
python -m classic.test --random-policy --episodes 1 --max-steps 10 --no-render
```

开始训练：

```bash
python classic/train.py --algo sac --robot ur5_cxy --run-name exp_sac --timesteps 500000 --device cuda --no-render
```

测试模型：

```bash
python classic/test.py --algo sac --robot ur5_cxy --run-name exp_sac --episodes 3 --render
```

Reproduce zeroarm legacy end-effector velocity semantics:

```bash
python classic/train.py --algo sac --robot zero_robotiq --run-name exp_zero_legacy --legacy-zero-ee-velocity --no-render
```
