## 2. 看环境层

- `_sample_target_position()`：课程学习如何采样目标
- `_action_to_control()`：策略输出如何映射成 MuJoCo 控制量
- `_compute_reward()`：奖励如何拆成多个分量


```python
from ur5_reach_env import UR5ReachEnv

env = UR5ReachEnv()
obs, info = env.reset(seed=42)
print('obs shape:', obs.shape)
print('info:', info)
env.close()
```
